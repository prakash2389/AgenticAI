```python

from crewai import Agent
```

```python
#Create a senior bog content researcher

bog_researcher = Agent(name="Blog Researcher",
                       role ="Blog searcher from Youtube Videos",
                       goal = "get the relavent video content for the topic{topic} from Youtube channel",
                       verbose=True,
                       memory=True,
                       backstory ="Expert in understanding You tube channel content and produces new content",
                       tools=[]
)
```

```python
blog_writer = Agent(role="Blog writer",
                  goal="Writes blogs based topic{topic}",
                  verbose=True,
                  memory=True,
                  backstory ="Expert in writing the blogs based on understanding relavent topic",
                  tools=[]
                  )
```

```python
ytool = YoutubeChannelSearchTool(youtube_channel_handle='@codebasics')
```

```python
Task1 = Task(
    agent=bog_researcher,
    description = "As blog_researcher, go through the youtube page and check all the content about the topic{topic} and get content",
    expected_output="Get the topics discussed in the channel",
    tools =[ytool],
    async_execution = False
)
```

```python
Task2 = Task(
    agent=blog_writer,
    description = "As Task writer, write some thing about topic{topic} and create content",
    expected_output="A 3 paragraph on topics discussed about the topic",
    async_execution = False,
    tools=[ytool],
    output_file = "new_blog.md"
)
```

```python
from crewai import Crew,Process
```

```python
Crewrun = Crew(
    agents=[bog_researcher,blog_writer],
    tasks=[Task1,Task2],
    process = Process.sequential,
    verbose=True,
    debug=True,
    planning=True
)
```

```python
result = Crewrun.kickoff(inputs={"topic": "machine Learning"})
print(result)
```
